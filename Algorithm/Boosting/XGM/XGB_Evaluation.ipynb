```python
import pandas as pd
import numpy as np
from pickle import load
```

```python
data = pd.read_csv("../../../Datasets/titanic.csv")
y=data["alive"]
data.drop(['alive'], axis = 1,inplace=True)
```

# Spliting the Categorical and Numerical

```python
Categoricaldata=data.select_dtypes(include=['object','bool'])
Categoricaldata
```

```python
Nuericaldata=data.select_dtypes(exclude=['object','bool'])
Nuericaldata
```

# Load the Imputer Model

```python
imputer_model = load(open('imputer.pkl', 'rb'))
Numericaldata1 = pd.DataFrame(imputer_model.transform(Nuericaldata))
Numericaldata1.columns = Nuericaldata.columns
Numericaldata1
```

```python
imputer1_model = load(open('imputer1.pkl', 'rb'))
Categoricaldata1 = pd.DataFrame(imputer1_model.transform(Categoricaldata))
Categoricaldata1.columns = Categoricaldata.columns
Categoricaldata1
```

# Encoding

```python
encoder_model = load(open('encoder.pkl', 'rb'))
encoder_model
```

```python
Categoricaldata2=pd.DataFrame(encoder_model.transform(Categoricaldata1))

#Getting Columns for Categorical_columns
cat_col = pd.get_dummies(Categoricaldata1)
Categoricaldata2.columns = cat_col.columns
Categoricaldata2

Categoricaldata2
```

```python
data = pd.concat([Numericaldata1,Categoricaldata2],axis=1)
data
```

# Loading the Scalar Model

```python
sclar_model = load(open('sclar.pkl', 'rb'))
sclar_model

data1 = pd.DataFrame(sclar_model.transform(data))
data1.columns=data.columns
data1
```

# Loading The XGM model

```python
xgmodel_model = load(open('xgmodel.pkl', 'rb'))
xgmodel_model
```

```python
preds = pd.DataFrame(xgmodel_model.predict(data1))
preds
```
