```python
import pandas as pd
import numpy as np
from pickle import dump
```

```python
data = pd.read_csv("../../../Datasets/titanic.csv")
y=data["alive"]
data.drop(['alive'], axis = 1,inplace=True)
```

# Spliting Categorical and Numerical Columns 

```python
Categoricaldata=data.select_dtypes(include=['object','bool'])
Categoricaldata
```

```python
Nuericaldata=data.select_dtypes(exclude=['object','bool'])
Nuericaldata
```

# Imputer method

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan,strategy ='most_frequent')
imputer.fit(Nuericaldata)

Numericaldata1 = pd.DataFrame(imputer.transform(Nuericaldata))
Numericaldata1.columns = Nuericaldata.columns
Numericaldata1

```

```python
#Save the model
dump(imputer, open('imputer.pkl', 'wb'))
```

```python
imputer = SimpleImputer(missing_values = np.nan,strategy ='most_frequent')
imputer.fit(Categoricaldata)

Categoricaldata1 = pd.DataFrame(imputer.transform(Categoricaldata))
Categoricaldata1.columns = Categoricaldata.columns
Categoricaldata1

```

```python
#Save the model
dump(imputer, open('imputer1.pkl', 'wb'))
```

# One_Hot_Encoding

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
encoder.fit(Categoricaldata1)
Categoricaldata2=pd.DataFrame(encoder.transform(Categoricaldata1))

#Save the model
dump(encoder, open('encoder.pkl', 'wb'))

#Getting Columns for Categorical_columns
cat_col = pd.get_dummies(Categoricaldata1)
Categoricaldata2.columns = cat_col.columns
Categoricaldata2
```

```python
data = pd.concat([Numericaldata1,Categoricaldata2],axis=1)
data
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
sclar = StandardScaler()
sclar.fit(data)

#Save the model
dump(sclar, open('sclar.pkl', 'wb'))

data1 = pd.DataFrame(sclar.transform(data))
data1.columns=data.columns
data1
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size = 0.2, random_state = 123)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
xg_cl = xgb.XGBClassifier(objective='binary:logistic', n_estimators = 10, seed = 123)
xg_cl.fit(X_train,y_train)

#Save the model
dump(xg_cl, open('xgmodel.pkl', 'wb'))
```

```python
preds = pd.DataFrame(xg_cl.predict(X_test))
preds
```

```python
y_test=pd.DataFrame(y_test)
y_test
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, preds)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, preds)
```
